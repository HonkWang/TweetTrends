```python
import json
import pandas as pd
import matplotlib.pyplot as plt
from Tkinter import Tk
from tkFileDialog import askopenfilename

Tk().withdraw()
tweets_data_path = askopenfilename()
# tweets_data_path = '../twitter.txt'

tweets_data = []
tweets =open(tweets_data_path,'r')

for line in tweets:
    try:
        tweet = json.loads(line)
        tweets_data.append(json.loads(line))
    except:
        continue
```

```python
tweetsDF = pd.DataFrame()

# map(f, iterable)
# is basically equivalent to:
# [f(x) for x in iterable]

tweetsDF['text'] = map(lambda tweet: tweet['text'], tweets_data)
tweetsDF['lang'] = map(lambda tweet: tweet['lang'], tweets_data)
tweetsDF['country'] = map(lambda tweet: tweet['place']['country']  if tweet['place']!= None else None, tweets_data)
```

```python
tweetsDF
```

```python
import matplotlib.pyplot as plt
tweets_by_lang = tweetsDF['lang'].value_counts()

figure, ax = plt.subplots()
ax.tick_params(axis='x', labelsize=15)
ax.tick_params(axis='y', labelsize=10)
ax.set_xlabel('Languages', fontsize=15)
ax.set_ylabel('Number of tweets' , fontsize=15)
ax.set_title('Top 5 languages', fontsize=15, fontweight='bold')
tweets_by_lang[:5].plot(ax=ax, kind='bar', color='red') #top five language

figure
```
